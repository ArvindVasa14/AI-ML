```python
import numpy as np
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
arr1= np.arange(2,17,2)

arr1
```

```python
mean= arr1.mean()
mean
```

```python
std= arr1.std()
std
```

```python
def z_score(num):
    return ((num-mean)/(std))
```

```python
arr2= np.array([z_score(i) for i in arr1])
arr2
```

```python
arr2.mean()
```

```python
arr2.std()
```

```python
arr2
```

```python
sns.kdeplot(arr2)
sns.kdeplot(arr1)
```
