```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import itertools
from sklearn import tree
```

### Label encoding

```python
def preprocess(df):
    dup_df= df.copy()
    le= preprocessing.LabelEncoder()
    
    dup_df['job']= le.fit_transform(dup_df['job'])
    dup_df['marital']= le.fit_transform(dup_df['marital'])
    dup_df['education']= le.fit_transform(dup_df['education'])
    dup_df['default']= le.fit_transform(dup_df['default'])
    dup_df['housing']= le.fit_transform(dup_df['housing'])
    dup_df['loan']= le.fit_transform(dup_df['loan'])
    dup_df['deposit']= le.fit_transform(dup_df['deposit'])
    dup_df['contact']= le.fit_transform(dup_df['contact'])
    dup_df['month']= le.fit_transform(dup_df['month'])
    dup_df['poutcome']= le.fit_transform(dup_df['poutcome'])
    
    return dup_df
```

```python
processed_data= preprocess(data)
processed_data.head()
```

```python
x= processed_data.drop(['deposit'],axis=1)
y= processed_data['deposit']
```

### Splitting data into train and test

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.2)
```

### Building Decision Tree

```python
max_depth= range(4,10)
min_samples_split= [60,70,80,90,100]
min_samples_leaf= [60,70,80,90,100]
max_scr=0
opt_depth=0
opt_split=0
opt_leaf=0
```

```python
permutations= list(itertools.product(max_depth,min_samples_split,min_samples_leaf))
len(permutations)
```

```python
def build_dt(dep,split,leaf):
    model= DecisionTreeClassifier(random_state=1, min_samples_leaf=leaf, min_samples_split=split,max_depth=dep,criterion='gini')
    model.fit(x_train,y_train)
    md_score= model.score(x_test,y_test)
    
    return md_score*100,model
```

```python
for comb in permutations:
    md_score, model= build_dt(comb[0],comb[1],comb[2])
#     print(md_score>max_scr)
    if(md_score>max_scr):
        print(comb)
        opt_depth= comb[0]
        opt_split= comb[1]
        opt_leaf= comb[2]
        max_scr= md_score
        tre= model
```

```python
print(("max efficiency: {}, at max_depth: {}, min_samples_split: {}, min_samples_leaf: {}").format(max_scr, opt_depth, opt_split, opt_leaf))
fig, ax = plt.subplots(figsize=(45, 20))
tree.plot_tree(model, feature_names=list(x.columns),filled=True, ax=ax, fontsize=10)
plt.show()
```
